# **TASK - 1**

1.  Find the linear regression equation for mileage vs price.
2.  Chart the original data and the equation on the chart.
3.  Find the equation's  R2  score (use the .score method) to determine whether the equation is a good fit for this data. (0.8 and greater is considered a strong correlation.)

```python
x = df[['Mileage']].values
y = df['Price'].values
```

```python
model = linear_model.LinearRegression()
model.fit(x, y)

print("intercept:", model.intercept_)
print("co_efficient:", model.coef_)
y_pred = model.predict(x)
```

```python
plt.style.use("ggplot")
sns.scatterplot(x="Mileage", y="Price", data= df)
plt.plot(x, y_pred, color="Black")
plt.xlabel("Mileage")
plt.ylabel("Price")
plt.title("Mileage Vs Price")
```

```python
r2 = model.score(x,y)
print("R Squared:", r2)
```

The Value of **R squared** by using score method is **0.0204** which is **very less than 0.8**, so the **equation for Mileage Vs Price is not correlated** with the data 

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=0)
X_train_poly = poly.fit_transform(x)
poly_fit = LinearRegression().fit(X_train_poly, y)
r0 = poly_fit.score(X_train_poly, y)
print("degree=0, r_squared value:", r0)

poly = PolynomialFeatures(degree=1)
X_train_poly = poly.fit_transform(x)
poly_fit = LinearRegression().fit(X_train_poly, y)
r1 =poly_fit.score(X_train_poly, y)
print("degree=1, r_squared value:", r1)

poly = PolynomialFeatures(degree=2)
X_train_poly = poly.fit_transform(x)
poly_fit = LinearRegression().fit(X_train_poly, y)
r2 =poly_fit.score(X_train_poly, y)
print("degree=2, r_squared value:", r2)

poly = PolynomialFeatures(degree=3)
X_train_poly = poly.fit_transform(x)
poly_fit = LinearRegression().fit(X_train_poly, y)
r3=poly_fit.score(X_train_poly, y)
print("degree=3, r_squared value:", r3)

poly = PolynomialFeatures(degree=4)
X_train_poly = poly.fit_transform(x)
poly_fit = LinearRegression().fit(X_train_poly, y)
r4= poly_fit.score(X_train_poly, y)
print("degree=4, r_squared value:", r4)

poly = PolynomialFeatures(degree=5)
X_train_poly = poly.fit_transform(x)
poly_fit = LinearRegression().fit(X_train_poly, y)
r5=poly_fit.score(X_train_poly, y)
print("degree=5, r_squared value:", r5)
```

As the **degree increases** the value **R Squared has also increases ** simultaneously

# **TASK- 2**

1. Use mileage, cylinders, liters, doors, cruise, sound, and leather to find the linear regression equation.
2. Find the equation's  R2  score (use the .score method) to determine whether the equation is a good fit for this data. (0.8 and greater is considered a strong correlation.)
3. Find the combination of the factors that is the best predictor for price.

```python
attribute = ['Mileage', 'Cylinder',	'Liter',	'Doors',	'Cruise',	'Sound',	'Leather']
x1 = df[attribute].values
y1 = df["Price"].values
```

```python
regressor = linear_model.LinearRegression()
regressor.fit(x1, y1)

print("intercept:", regressor.intercept_)
print("co_efficient:", regressor.coef_)
y_predict = regressor.predict(x1)
```

```python
print("Each attribute with its coefficient:", (list(zip(attribute, regressor.coef_))), sep="\n")
```

```python
sns.barplot(y_predict)
```

```python
plt.plot(x1, y1, "*", color="green")
plt.plot( x1, y_predict, "+", color="pink")
```

```python
r_2 = regressor.score(x1, y1)
print("R squared value:", r_2)
```

The value of **R Squared Value** for attribute vs price is **0.446**, **less than 0.8** so the equation is **not correlated** or weak fit 
with the data. 

### Find the combination of the factors that is the best predictor for price

```python
def scatter(a,fig):
    plt.subplot(5,2,fig)
    plt.scatter(df[a],df['Price'], color="limegreen")
    plt.title(a+' vs Price')
    plt.ylabel('Price')
    plt.xlabel(a)

plt.figure(figsize=(10,20))

scatter('Mileage', 1)
scatter('Cylinder', 2)
scatter('Liter', 3)
scatter('Doors', 4)
scatter('Cruise', 5)
scatter('Sound', 6)
scatter('Leather', 7)

plt.tight_layout()
```

         2. Whereas Liter has **better significant** to predict the Price when compared to Mileage.
